### Line pots (series/dataframe) 

#### Question: Plot a line graph of immigration from Haiti using df.plot() 

```python
haiti = df_can.loc['Haiti', years]
haiti.head()
```

```python
haiti.plot()
```

```python
# Add labels then plot

haiti.index = haiti.index.map(int)
haiti.plot(kind='line')

plt.title('Immigration from Haiti')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')

plt.show()
```

#### Question: Compare the number of immigrants from India and China from 1980 to 2013.

```python
# 1 - Get the dataset for China and India, and display the dataframe
china_india_df = df_can.loc[['China', 'India'], years]
china_india_df.head()
```

```python
# 2 - Plot the graph using plot()
china_india_df.plot()
```

```python
# Transpose data
china_india_df = china_india_df.transpose()
china_india_df.head()
```

```python
ci_df = china_india_df

ci_df.index = ci_df.index.map(int)
ci_df.plot(kind='line')

plt.title('Comparison between immigration from China and India')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')
plt.show()
```

#### Compare the trend of the top 5 countries that contribute the most to immigration in Canada

```python
# View top 5 countries with the most immigrants
df_can.sort_values(by=['Total'], ascending=False)
```

```python
top5_df = df_can.loc[['India', 'China', 'United Kingdom of Great Britain and Northern Ireland', 'Philippines', 'Pakistan'], years]
top5_df
```

```python
top5_df.plot()
```

```python
top5_df = top5_df.transpose()
top5_df
```

```python
top5_df.index = top5_df.index.map(int)
top5_df.plot(kind='line')

plt.title('Comparison between the top 5 countries with the most number of immigrants to Canada')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')

plt.show()
```
